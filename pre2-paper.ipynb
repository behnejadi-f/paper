```python
import numpy as np
from scipy import misc
import glob
```

```python
TCClass = 0.0
TSClass = 1.0
imgDimension = 100
```

```python
f_handle = open('data.row\\ASDDataset_Augmented.csv', 'a')#The CSV file where data records are appended to
```

```python
#Formatting the header row of the CSV file
headerRow = ""
for i in range (0,(imgDimension*imgDimension)):
    headerRow = headerRow + "Pixel" + str(i)+ ","
headerRow = headerRow+"Label"+"\n"
f_handle.write(headerRow)
```

```python
import matplotlib.pyplot
import numpy as np
from skimage.transform import resize
from PIL import Image


#Processing the TC images
print("Processing TC images...")
for file in glob.iglob('data.row\\image\\TCImages\\**\\*.png', recursive=True):
    print("Current File:", file)
    img = matplotlib.pyplot.imread(file)
    img = img[: ,:, :3]#Excluding the Alpha channel
    img = resize(img, (imgDimension, imgDimension))
    #img = np.array(Image.fromarray(img).resize((imgDimension, imgDimension)))
        
#xnr = np.array(Image.fromarray(xn).resize((200, xn.shape[1])))

    img = (img / 255.0)
    img = img.reshape(-1, 3)

    red = img[:,0]
    green = img[:,1]
    blue = img[:,2]
    gray = 0.299 * red + 0.587 * green + 0.114 * blue #Converting into grayscale
    gray = np.append(gray, TCClass)
    gray = gray.reshape(1, (imgDimension*imgDimension)+1)
    np.savetxt(f_handle, gray, fmt='%.4g', delimiter=",")

print("TC Done.")

#Same process for TS images
for file in glob.iglob('data.row\\image\\TSImages\\**\\*.png', recursive=True):
    print("Current File:", file)
    img = matplotlib.pyplot.imread(file)
    img = img[:,:,:3]#Excluding the Alpha channel
    img = resize(img, (imgDimension, imgDimension))
    #img = np.array(Image.fromarray(img).resize((imgDimension, imgDimension)))

    img = (img / 255.0)
    img = img.reshape(-1, 3)
    red = img[:,0]
    green = img[:,1]
    blue = img[:,2]
    gray = 0.299 * red + 0.587 * green + 0.114 * blue
    gray = np.append(gray, TSClass)
    gray = gray.reshape(1, (imgDimension*imgDimension)+1)

    np.savetxt(f_handle, gray, fmt='%.4g',delimiter=",")
f_handle.close()
```
